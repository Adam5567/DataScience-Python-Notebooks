```python
data = sqlite3.connect('database.sqlite')
```

```python
#All of the tables
tables = pd.read_sql("""SELECT * FROM sqlite_master WHERE type='table';""", data)
tables
```

```python
#Countries in the data
countries = pd.read_sql("""SELECT * FROM Country;""", data)
countries
```

```python
#Players in the data
players = pd.read_sql("""SELECT * FROM Player;""", data)

players
```

```python
#Lets find out how many players are over 190cm.

tall_players = pd.read_sql("""SELECT player_name, height FROM Player WHERE height > 190;""", data)

tall_players
```

```python
#Look at the League table, we see that the ID and Country_ID are the same. 
#The Country_ID is actually a foreign key of the Country table
league = pd.read_sql("""SELECT * FROM League;""", data)
league
```

```python
#If we wanted to, we could combine the League and Country tables:

CountryLeagues = pd.read_sql("""SELECT * FROM League JOIN Country ON Country.id = League.country_id;""", data)
CountryLeagues
```
